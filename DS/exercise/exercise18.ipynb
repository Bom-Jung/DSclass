```python
import requests
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%A9%94%ED%83%80%EB%B2%84%EC%8A%A4'
res= requests.get(url)


html_doc = res.text

print(html_doc)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_doc,'html.parser')
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
news = soup.find_all("a",{"class" : "news_tit"})

print(news)
```

```python
# First news
news[0]
```

```python
# Get a url of the first news
news[0].get('href')
```

```python
# Get a title
news[0].get('title')
```
